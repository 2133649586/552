```python
from sklearn.multioutput import ClassifierChain
from collections import Counter
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import jaccard_similarity_score

class_name = ['Family','Genus','Species']



for i in class_name:
    svc = LinearSVC(penalty='l1', multi_class='ovr', dual=False)
    model_ovr = OneVsRestClassifier(svc)
    model_ovr.fit(X_train, y_train[i])
    pred_bv = model_ovr.predict(X_test)
    jaccard_score = jaccard_similarity_score(y_test[i], pred_bv)
    
    chains = [ClassifierChain(svc, order='random',random_state=b) for b in range(10)]

    predict_chains = []
    chain_jaccard = []
    
    y_train_np = np.array(y_train[i])
    y_test_np = np.array(y_test[i])
    X_test_np = np.array(X_test)
    X_train_np = np.array(X_train)
    
    le = LabelEncoder()
    le.fit(y_train_np)
    le_y_train = le.transform(y_train_np)
    le_y_train = np.array(le_y_train).reshape(-1,1)
    
    le = LabelEncoder()
    le.fit(y_test_np)
    le_y_test = le.transform(y_test_np)
    le_y_test = np.array(le_y_test).reshape(-1,1)

    
#     print(le_y_train)
#     print(le_y_test)

    
    
    for chain in chains:
        chain.fit(X_train_np, le_y_train)
        predict_pre_chain = chain.predict(X_test)

        
        predict_chains.append(predict_pre_chain)
        chain_jaccard.append(jaccard_similarity_score(le_y_test,predict_pre_chain >= .5))
    
    predict_chains_ensemble = np.array(predict_chains).mean(axis=0)
    ensemble_jaccard_score = jaccard_similarity_score(le_y_test,predict_chains_ensemble >= 0.5)
    model_scores = chain_jaccard+[ensemble_jaccard_score]
    model_names = ('Chain 1','Chain 2','Chain 3','Chain 4','Chain 5','Chain 6','Chain 7','Chain 8',\
              'Chain 9','Chain 10','Ensemble')

    x_pos = np.arange(len(model_names))
    
    
    fig, ax = plt.subplots(figsize=(7, 4))
    ax.grid(True)
    ax.set_title(f'Classifier Chain for {i}')
    ax.set_xticks(x_pos)
    ax.set_xticklabels(model_names, rotation='vertical')
    ax.set_ylabel('Jaccard Similarity Score')
    ax.set_ylim([min(model_scores) * 0.9, max(model_scores) * 1.1])
    colors = ['r'] + ['b'] * len(chain_jaccard) + ['g']
    ax.bar(x_pos, model_scores, alpha=0.5, color=colors)
    plt.tight_layout()
    plt.show()
    
        
    
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix
from sklearn.svm import LinearSVC
from sklearn.grid_search import GridSearchCV
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegressionCV
from sklearn.preprocessing import label_binarize



#params = {'C':[10**(-8),10**(-1),10**(1)]}
params = {'C':[10**i for i in range(-3,7)]}

    
LSVC = LinearSVC(penalty='l1', multi_class='ovr', dual=False)
model_aiii = GridSearchCV(LSVC, params, cv=10,scoring='accuracy')
model_aiii.fit(X_train,y_train['Family'])
predict_aiii = model_aiii.predict(X_test)

hamming_score = hamming_loss(y_test['Family'], predict_aiii)
exact_match_score = accuracy_score(y_test['Family'], predict_aiii)
precision_score = metrics.precision_score(y_test['Family'], predict_aiii, average='micro')
recall_score = metrics.recall_score(y_test['Family'], predict_aiii, average='micro')


print('Family')
#print(model_aiii.best_params_)
print('hamming socre: ', hamming_score)
print('exact match score: ', exact_match_score)
print('precision score: ', precision_score)
print('recall score: ', recall_score)
print(' ')

cmatrix = confusion_matrix(y_test['Family'], predict_aiii)
#plt.matshow(cmatrix, cmap=plt.cm.gray)
row_sums = cmatrix.sum(axis= 1, keepdims= True)
normal_cm = cmatrix/row_sums
np.fill_diagonal(normal_cm, 0)
plt.matshow(normal_cm, cmap=plt.cm.gray)



le = LabelEncoder()
le.fit(y_train['Family'])
le_y_train = le.transform(y_train['Family'])
le_y_train = np.array(le_y_train).reshape(-1,1)
le_y_test = le.transform(y_test['Family'])
le_y_test = np.array(le_y_test).reshape(-1,1)

y_train_label = label_binarize(le_y_train, classes=[0,1,2,3])
y_test_label= label_binarize(le_y_test, classes=[0,1,2,3])

classifier = OneVsRestClassifier(GridSearchCV(LSVC, params, cv=10,scoring='accuracy'))
#classifier = OneVsRestClassifier(LogisticRegressionCV(penalty='l1',cv=StratifiedKFold(5),solver='liblinear'))
y_score = classifier.fit(X_train, y_train_label).decision_function(X_test)

n_classes = y_train_label.shape[1]
fpr = {}
tpr = {}
ROCAUC = {}

for i in range(0,n_classes):
    fpr[i],tpr[i],a = roc_curve(y_test_label[:,i],y_score[:,i])
    ROCAUC[i] = auc(fpr[i],tpr[i])
    
plt.figure()

colors = ['red','green','yellow','blue','orange','gray']
for i in range(0,n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i], lw=2,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, ROCAUC[i]))



plt.plot([0, 1], [0, 1], '--', lw=2)
plt.xlim([-0.05, 1.0])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import hamming_loss
from collections import Counter
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from scipy.spatial import distance


row_file = pd.read_csv('/Users/liyifan/Desktop/551552/homework/homework5/Frogs_MFCCs.csv')
row_label_data = row_file[['Family','Genus','Species']]
row_data = row_file[variable_name]

hamming_loss_list = []
hamming_distance = []
for n in range(1,51):
    row_file = pd.read_csv('/Users/liyifan/Desktop/551552/homework/homework5/Frogs_MFCCs.csv')
    row_label_data = row_file[['Family','Genus','Species']]
    row_data = row_file[variable_name]
    silh_score_dict = {}
    
    
    for i in range(2,51):    
    #for i in range(2,15):
        kmean = KMeans(n_clusters=i, random_state=n)
        kmean.fit(row_data)
        kmean_label = kmean.fit_predict(row_data)
        silh_score = silhouette_score(row_data, kmean_label)
        silh_score_dict[i]=silh_score

    best_k = max(silh_score_dict.keys(),key=(lambda x:silh_score_dict[x]))

    
    
    best_kmean = KMeans(n_clusters=best_k, random_state=n)
    new_label = best_kmean.fit_predict(row_data)
    add_newlabel_data = row_file.drop(['RecordID'], axis=1)
    add_newlabel_data.insert(25,'Newlabel',new_label)
    add_newlabel_data_np = np.array(add_newlabel_data)
    
    
    cluster_dict = {}
    family_dict = {}
    genus_dict = {}
    species_dict = {}
    new_label_1 = new_label
    new_label_2 = new_label
    new_label_3 = new_label
    label_dict = {'Family':[],'Genus':[],'Species':[]}
    
    print(f'the {n}th rum:')
    
    '''Family'''
    Family_list = []
    for i in range(0,best_k):
        select_data = add_newlabel_data_np[np.where(add_newlabel_data_np[:,25]==i)]
        Family_data = select_data[:,22]
        counter_Family_data = Counter(Family_data).most_common()
        family_dict[f'the {i} cluster'] = counter_Family_data[0][0]
        
        Family_list.append(counter_Family_data[0][0])
#         for b in range(0,len(new_label)):
#             if label_dict['Family'][b]==i:
#                 label_dict['Family'][b]=str(counter_Family_data[0][0])
    print('Family', Family_list)
      
    for b in range(0,len(new_label)):
        label_dict['Family'].append(Family_list[new_label[b]])

    
    '''Genus'''
    Genus_list = []
    for i in range(0,best_k):
        select_data = add_newlabel_data_np[np.where(add_newlabel_data_np[:,25]==i)]
        Genus_data = select_data[:,23]
        counter_Genus_data = Counter(Genus_data).most_common()
        genus_dict[f'the {i} cluster'] = counter_Genus_data[0][0]
        
        Genus_list.append(counter_Genus_data[0][0])
#         for b in range(0,len(label_dict['Genus'])):
#             if label_dict['Genus'][b]==i:
#                 label_dict['Genus'][b] = counter_Genus_data[0][0]

    print('Genus', Genus_list)

    for b in range(0,len(new_label)):
        label_dict['Genus'].append(Genus_list[new_label[b]])

    
    '''Species'''
    Species_list = []
    for i in range(0,best_k):
        select_data = add_newlabel_data_np[np.where(add_newlabel_data_np[:,25]==i)]
        Species_data = select_data[:,24]
        counter_Species_data = Counter(Species_data).most_common()
        species_dict[f'the {i} cluster'] = counter_Species_data[0][0] 
        
        Species_list.append(counter_Species_data[0][0])
#         for b in range(0,len(label_dict['Species'])):
#             if new_label==i:
#                 new_label['Species'].append(counter_Species_data[0][0]) 


    print('Species', Species_list)
    print('\n')
    print('\n')
        
    for b in range(0,len(new_label)):
        label_dict['Species'].append(Species_list[new_label[b]])
        
    
    cluster_dict['family'] = family_dict
    cluster_dict['genus'] = genus_dict
    cluster_dict['species'] = species_dict
        
    hamming_loss_list.append((hamming_loss(row_file['Family'], label_dict['Family'])+hamming_loss(row_file['Genus'],\
                            label_dict['Genus'])+hamming_loss(row_file['Species'], label_dict['Species']))/3)
    hamming_distance.append((distance.hamming(row_file['Family'], label_dict['Family'])+distance.hamming(row_file['Genus'],\
                            label_dict['Genus'])+distance.hamming(row_file['Species'], label_dict['Species']))/3)
hamming_loss_list = np.array(hamming_loss_list)
hamming_distance = np.array(hamming_distance)

standard_deviation_of_Hamming_Distances = np.std(hamming_distance)
average_Hamming_distance = np.mean(hamming_distance)
Hamming_loss = np.mean(hamming_loss_list)
Hamming_score = 1-Hamming_loss
print('\n')
print('\n')
print('\n')
print('average Hamming distance is:',average_Hamming_distance,'\n Hamming score is:',Hamming_score,'\n Hamming loss is:',Hamming_loss)
print('standard deviation of Hamming Distances is:',standard_deviation_of_Hamming_Distances) 

        
#ab_dataset = dataset[np.where(dataset[:,6]=='Abnormal')]
```
